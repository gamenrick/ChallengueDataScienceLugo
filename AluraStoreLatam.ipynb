#1. Análisis de facturación



Obtener los ingresos de cada producto acorde a la cantidad de unidades vendidas.

```python
tienda1['Facturación'] = tienda1['Cantidad de cuotas'] * tienda1['Precio']
tienda2['Facturación'] = tienda2['Cantidad de cuotas'] * tienda2['Precio']
tienda3['Facturación'] = tienda3['Cantidad de cuotas'] * tienda3['Precio']
tienda4['Facturación'] = tienda4['Cantidad de cuotas'] * tienda4['Precio']
```

Obtener un total de valor de cada uno de los productos vendidos.

```python
facturacionTienda1 = tienda1['Facturación'].sum()
facturacionTienda2 = tienda2['Facturación'].sum()
facturacionTienda3 = tienda3['Facturación'].sum()
facturacionTienda4 = tienda4['Facturación'].sum()
```

```python
print(f'Facturacion Mensual de la Tienda 1: {facturacionTienda1}')
print(f'Facturacion Mensual de la Tienda 2: {facturacionTienda2}')
print(f'Facturacion Mensual de la Tienda 3: {facturacionTienda3}')
print(f'Facturacion Mensual de la Tienda 4: {facturacionTienda4}')
```

# 4. Productos más y menos vendidos

```python
ProductosVendidosTienda1 = tienda1['Producto'].value_counts().to_dict()
ProductosVendidosTienda2 = tienda2['Producto'].value_counts().to_dict()
ProductosVendidosTienda3 = tienda3['Producto'].value_counts().to_dict()
ProductosVendidosTienda4 = tienda4['Producto'].value_counts().to_dict()
```

```python
import pandas as pd

# ... (Your existing code to load data into tienda, tienda2, tienda3, tienda4) ...

# Create a list to store individual product records
all_products = []

for i, tienda_data in enumerate([tienda1, tienda2, tienda3, tienda4]):
    productos_vendidos = tienda_data['Producto'].value_counts().to_dict()
    for product, count in productos_vendidos.items():
        all_products.append({'Tienda': f'Tienda {i+1}', 'Producto': product, 'Cantidad': count})

# Create a pandas DataFrame from the product records
df = pd.DataFrame(all_products)

# Pivot the DataFrame to create a table with products as columns
product_table = df.pivot_table(index='Tienda', columns='Producto', values='Cantidad', fill_value=0)

# Display the formatted table
print(product_table)
```

```python
# Find and display the most and least sold products for each store
print("\nProductos Más y Menos Vendidos por Tienda:\n")
for tiendaSelec in product_table.index:
    most_sold = product_table.loc[tiendaSelec].idxmax()
    least_sold = product_table.loc[tiendaSelec].idxmin()
    print(f"{tiendaSelec}:")
    print(f"  Más Vendido: {most_sold}")
    print(f"  Menos Vendido: {least_sold}\n")
```

```python
import pandas as pd

# ... (Your existing code to load data into tienda, tienda2, tienda3, tienda4) ...

def calculate_average_shipping_cost(df):
    """Calculates the average shipping cost, handling potential errors."""
    try:
        # Attempt to convert the 'Costo de envío' column to numeric
        df['Costo de envío'] = pd.to_numeric(df['Costo de envío'], errors='coerce')

        # Calculate the mean of the 'Costo de envío' column
        average_cost = df['Costo de envío'].mean()

        return average_cost
    except KeyError:
        print("Error: 'Costo de envío' column not found in the DataFrame.")
        return None
    except TypeError as e:
        print(f"TypeError encountered: {e}")  # Print the specific error
        return None

# Calculate and print the average shipping cost for each store
costoEnvioPromedio = calculate_average_shipping_cost(tienda1)
costoEnvioPromedio2 = calculate_average_shipping_cost(tienda2)
costoEnvioPromedio3 = calculate_average_shipping_cost(tienda3)
costoEnvioPromedio4 = calculate_average_shipping_cost(tienda4)

if costoEnvioPromedio is not None:
    print(f'costo de envio promedio de la Tienda 1: {costoEnvioPromedio}')
if costoEnvioPromedio2 is not None:
    print(f'costo de envio promedio de la Tienda 2: {costoEnvioPromedio2}')
if costoEnvioPromedio3 is not None:
    print(f'costo de envio promedio de la Tienda 3: {costoEnvioPromedio3}')
if costoEnvioPromedio4 is not None:
    print(f'costo de envio promedio de la Tienda 4: {costoEnvioPromedio4}')
```

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
titulos = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
FacturacionDescuentoEnvio1 = facturacionTienda1 - costoEnvioPromedio
FacturacionDescuentoEnvio2 = facturacionTienda2 - costoEnvioPromedio2
FacturacionDescuentoEnvio3 = facturacionTienda3 - costoEnvioPromedio3
FacturacionDescuentoEnvio4 = facturacionTienda4 - costoEnvioPromedio4

FacturacionDescuentoEnvio = [
    FacturacionDescuentoEnvio1,
    FacturacionDescuentoEnvio2,
    FacturacionDescuentoEnvio3,
    FacturacionDescuentoEnvio4
]

calificacionTienda = [clasificacionPromedio, clasificacionPromedio2, clasificacionPromedio3, clasificacionPromedio4]
resultados = []

# Rango de facturación (ajusta según tus datos)
facturacion_min = 3000000000.0
facturacion_max = 3500000000.0

#Normalizar calificacionTienda y facturacionDescuentoEnvio a escala de 1 al 5

for descuento, calificacion in zip(FacturacionDescuentoEnvio, calificacionTienda):
    facturacion_normalizada = 1 + (descuento - facturacion_min) * 4 / (facturacion_max - facturacion_min)
    facturacion_normalizada = min(max(facturacion_normalizada, 1), 5)

    # Por ejemplo, ponderas 70% la facturación y 30% la calificación de clientes
    puntaje_total = 0.7 * facturacion_normalizada + 0.3 * calificacion
    resultados.append(puntaje_total)

# Graficar
plt.bar(titulos, resultados, color='skyblue')
plt.title('Evaluación Final')
plt.xlabel('Tiendas')
plt.ylabel('Calificación Final')
plt.ylim(0, 5)
plt.show()
```

# Informe Final: Recomendación de Cierre de Tienda


## Introducción

El presente informe tiene como objetivo determinar cuál de las cuatro tiendas del Sr. Juan debe ser vendida, con el fin de liberar capital y enfocarse en nuevas oportunidades de negocio. Para tomar esta decisión, se ha llevado a cabo un análisis integral considerando cinco aspectos clave: los ingresos totales por tienda, las categorías de productos más y menos vendidas, la calificación promedio otorgada por los clientes, los productos más y menos vendidos, y el costo promedio de envío. Este análisis permite evaluar el desempeño general de cada tienda desde una perspectiva financiera, operativa y de satisfacción del cliente.

## Desarrollo del Análisis

### 1. Ingresos Totales

Los ingresos mensuales por tienda muestran una clara diferencia en volumen de facturación:

Tienda 1: $3,422,717,700.00

Tienda 2: $3,337,189,200.00

Tienda 3: $3,243,816,900.00

Tienda 4: $3,098,827,200.00

La Tienda 4 presenta el ingreso más bajo, con una diferencia de más de 300 millones respecto a la Tienda 1.

### 2. Categorías de Productos Vendidas

Aunque todas las tiendas comparten una variedad similar de categorías, los volúmenes de venta difieren:

Tienda 1 lidera en ventas de Muebles y Electrónicos.

Tienda 3 tiene un buen equilibrio en varias categorías.

Tienda 4 presenta cifras competitivas, pero en general sus ventas están ligeramente por debajo de las demás en casi todas las categorías, en especial en Electrodomésticos e Instrumentos musicales.

### 3. Calificación Promedio de los Clientes

La satisfacción del cliente se mantiene alta en general, pero nuevamente la Tienda 4 queda en una posición menos favorable:

Tienda 3: 4.05 ⭐

Tienda 2: 4.04 ⭐

Tienda 4: 4.00 ⭐

Tienda 1: 3.98 ⭐

Aunque la Tienda 4 no tiene la peor calificación, tampoco sobresale en este aspecto.

### 4. Productos Más y Menos Vendidos

Tienda 1 destaca con productos como el Armario, mientras que su producto menos vendido es Auriculares con micrófono.

Tienda 4, en cambio, tiene como producto más vendido la Cama box, y como menos vendido la Guitarra eléctrica.

Esto refleja una menor diversidad de productos exitosos en la Tienda 4, con algunos productos de baja rotación significativa.

### 5. Costo de Envío Promedio

La Tienda 4 destaca positivamente aquí al tener el menor costo de envío:

Tienda 4: $23,459

Tienda 3: $24,805

Tienda 2: $25,216

Tienda 1: $26,018

Sin embargo, esta ventaja logística no compensa las debilidades en ingresos y ventas generales.


## Conclusión
Tras analizar todos los factores —ingresos, ventas por categoría, calificación de clientes, rotación de productos y costos logísticos— se concluye que la Tienda 4 es la mejor candidata para ser vendida.

A pesar de tener el menor costo de envío, la Tienda 4:

* Tiene la facturación más baja de las cuatro tiendas.

* Presenta una calificación promedio de clientes que no destaca sobre las demás.

* Muestra menores volúmenes de venta en la mayoría de las categorías.

* Tiene un producto menos vendido que evidencia una baja rotación de inventario.


Estos aspectos reflejan que la Tienda 4 no posee un desempeño competitivo en comparación con las otras tiendas. Venderla representaría una oportunidad estratégica para el Sr. Juan, permitiéndole liberar recursos que podrían destinarse a fortalecer las tiendas más rentables o explorar nuevas oportunidades de negocio.
